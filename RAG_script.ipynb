```python
import tiktoken

def get_num_tokens(text: str, encoding_name: str) -> int:
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(text))
    return num_tokens

get_num_tokens(full_document, "cl100k_base")
```

```python
from chromadb import Documents, EmbeddingFunction, Embeddings

class GeminiEmbeddingFunction(EmbeddingFunction):
  def __call__(self, input: Documents) -> Embeddings:
    model = 'models/embedding-001'
    title = "Custom query"
    embedding = genai.embed_content(
        model=model,
        content=input,
        task_type="retrieval_document",
        title=title
    )
    return embedding["embedding"]
  def embed_query(self, query: str) -> Embeddings:
    embedding = self.__call__(query)
    flat_embedding = [float(val) for sublist in embedding for val in sublist]
    return flat_embedding
  def embed_documents(self, documents: Documents) -> Embeddings:
    return self.__call__(documents)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=300,
    chunk_overlap=50,
)

splits = text_splitter.split_text(full_document)
```

```python
len(splits)
```

```python
from langchain.docstore.document import Document

# Convert text splits to Document objects
documents = [Document(page_content=split) for split in splits]

# Create the vector store
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=GeminiEmbeddingFunction(),
)
```

# **Retrieval**

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 4})
```

```python
docs = retriever.invoke("What is the gross salary of Managing Director?")
docs
```

# **Generation**

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
```

```python
# Create the prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Please respond to the user based only on the context"),
    ("user", "Question: {question}\nContext: {context}")
])

model = ChatGoogleGenerativeAI(model="gemini-1.5-flash")
```

```python
out_parser = StrOutputParser()

chain = prompt | model | out_parser

chain.invoke({"question": "What is the gross salary of Managing Director?", "context": docs})
```

# **Evaluation**

```python
import pandas as pd

ds = pd.read_excel("/content/QA_Pairs.xlsx")
ds.head(2)
```

```python
eval_questions = []
eval_ground_truths = []
eval_contexts = []

for _, row in ds.iterrows():
  eval_questions.append(row['question'])
  eval_ground_truths.append(row['text'])
  eval_contexts.append([row['context']])
```

```python
print("questions: ", type(eval_questions), type(eval_questions[0]))
print("ground truth: ", type(eval_ground_truths), type(eval_ground_truths[0]))
print("context: ", type(eval_contexts), type(eval_contexts[0]), type(eval_contexts[0][0]))
```

```python
eval_questions = eval_questions[:10]
eval_ground_truths = eval_ground_truths[:10]
eval_contexts = eval_contexts[:10]
```

```python
len(eval_questions), len(eval_ground_truths), len(eval_contexts)
```

```python
eval_docs = ""
for context in eval_contexts:
  eval_docs += context[0]
eval_docs
```

```python
eval_splits = text_splitter.split_text(eval_docs)
```

```python
eval_documents = [Document(page_content=split) for split in eval_splits]

eval_vectorstore = Chroma.from_documents(
    documents=eval_documents,
    embedding=GeminiEmbeddingFunction(),
)

eval_retriever = vectorstore.as_retriever(search_kwargs={"k": 4})
```

```python
eval_retriever.invoke("Who lives in the lake next to Hogwarts?")
```

```python
eval_chain = (
    {"context": eval_retriever, "question": RunnablePassthrough()}
    | prompt_hub_rag
    | model
    | out_parser
)
```

```python
eval_answers = []
eval_retrived_context = []

for question in eval_questions:
  eval_answers.append(eval_chain.invoke(question))
  eval_retrived_context.append([docs.page_content for docs in eval_retriever.invoke(question)])
```

```python
from ragas import evaluate
from datasets import Dataset
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    context_recall,
    context_precision
)

os.environ["OPENAI_API_KEY"] = "<api-key>"

eval_data = {
    "question": eval_questions,
    "ground_truth": eval_ground_truths,
    "contexts": eval_retrived_context,
    "answer": eval_answers
}

eval_dataset = Dataset.from_dict(eval_data)
```

```python
eval_result = evaluate(
    dataset=eval_dataset,
    metrics=[
        context_precision,
        context_recall,
        answer_relevancy,
        faithfulness
    ],
    raise_exceptions=False
)
```

```python
eval_df = eval_result.to_pandas()
```
